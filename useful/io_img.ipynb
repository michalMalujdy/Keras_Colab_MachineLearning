# Import image file

With specified name (*example_img.jpg*) among many selected

```
from google.colab import files
from io import BytesIO
from PIL import Image

filesDict = files.upload()
img = Image.open(BytesIO(filesDict['example_img.jpg']))
```



---



The first one from upload. Handy when uploaded single file.

```
from google.colab import files
from io import BytesIO
from PIL import Image

filesDict = files.upload()
img = Image.open(BytesIO(filesDict.values()[0]))
```



---



# Display the image with matplotlib

With default settings showing grid and axes numbers

```
import matplotlib.pyplot as plt

plt.imshow(img)
plt.show()
```



---



With no gridlines and axes numbers

```
import matplotlib.pyplot as plt

plt.figure().gca().grid(False)
plt.axis('off')

plt.imshow(img)
plt.show()
```
